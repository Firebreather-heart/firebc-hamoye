```python
y = df.pop('stabf')
x = df 
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
x_train, x_val, y_train,y_val = train_test_split(x,y, random_state=1, test_size=0.2)
```

```python
scl = StandardScaler()
scaled_t = scl.fit_transform(x_train)
scl_v = scl.fit_transform(x_val)
```

```python
x_train = pd.DataFrame(scaled_t, columns = x.columns)
x_val = pd.DataFrame(scl_v, columns = x.columns)
```

```python
y_train
```

```python
def trainmodel(model,  train_x, train_y,random_state=1,):
    mod = model(random_state = random_state)
    mod.fit(train_x,train_y)
    return mod
from sklearn.metrics import classification_report
def evalmodel(model, val_x,val_y):
    pred = model.predict(val_x)
    print(model, '\n', classification_report(pred, val_y))

class Dataflow:
    def __init__(self, trainx,trainy, valx,valy):
        self.models = [RandomForestClassifier, ExtraTreesClassifier, XGBClassifier, LGBMClassifier]
        self.x_t = trainx
        self.x_v = valx 
        self.y_t = trainy 
        self.y_v = valy
    
    def rollover(self):
        self.trained_models = []
        for model in self.models:
            self.trained_models.append(trainmodel(model, self.x_t, self.y_t))
        for model in self.trained_models:
            evalmodel(model, self.x_v, self.y_v)
```

```python
db = Dataflow(x_train, y_train, x_val, y_val)
```

```python
db.rollover()
```
